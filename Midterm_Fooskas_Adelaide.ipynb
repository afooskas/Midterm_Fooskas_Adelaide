```python
# Q1
# read the file
f=open('numbers.txt', 'r')
num1=f.read()
print(num1)

# counting frequencies
num2 = num1.replace('\n',',').split(',')
num3 = [int(i) for i in num2]
num3.sort
freq = {}
for number in num3:
    if number in freq:
        freq[number]+=1
    else:
        freq[number]=1
freq1=dict(sorted(freq.items()))
print(freq1)

# histogram
import matplotlib.pyplot as plt
import numpy as np
num3 = [key for key, val in freq.items() for _ in range(val)]
plt.hist(num3, bins=11, color="green")
plt.show()
# store results into json file
import json
jnum=json.dumps(freq1)
print(jnum)
```

```python
# Q2
# Importing Data
import pandas as pd
yup = open("zon.csv", "r")
print(yup.read())
```

```python
# Cleaning up and Reading Data
df=pd.read_csv("zon.csv")
df=df.drop(["Order ID", "Payment Instrument Type", "Purchase Order Number", "Ordering Customer Email", "Shipping Address Name", "Shipping Address Street 1", "Shipping Address Street 2", "Shipping Address City", "Shipping Address State", "Shipping Address Zip", "Order Status", "Carrier Name & Tracking Number", "Buyer Name", "Group Name"], axis=1)
df
```

```python
# Finding Monthly
import datetime
dfnew=df.groupby("Month_Num")["Total Charged"].describe()
print(dfnew)
```

```python
# Total Charged Analysis
df["Total Charged"].describe()
```

```python
# Orders per Month (Pie Chart)
mycolors=["orange", "dodgerblue","darkgoldenrod", "lightsteelblue"]
mylabels=["april", "may", "june", "july"]
plt1=dfnew.plot.pie(y="count",labels=mylabels,colors=mycolors, title="Orders per Month")
plt.legend(title="Month", bbox_to_anchor=(1.05,1.0),loc="upper left")
plt.show()
```

```python
# Average Charged per Month (Bar Chart)
dfnew.plot.bar(y="mean", color="royalblue")
plt.xlabel("Month")
plt.ylabel("Value")
plt.title("Average Charged per Month")
plt.show()
```

```python
# Compare Max and Min (Multi Line)
ax=plt.gca()
dfnew.plot(y="max", ax=ax, color = "darkturquoise")
dfnew.plot(y="min", ax=ax, color = "olivedrab")
plt.xlabel("Month")
plt.ylabel("Value")
plt.title("Min vs Max Spent Each Month")
plt.show()
```
